### Worker's Data

```python
# %pip install plotly
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import plotly.express as pe
import numpy as np
```

```python
pd.set_option('display.max_column', None)
mb = pd.read_csv('workers_data-1.csv')

C = mb.select_dtypes(include = ['object','category'])
D = mb.select_dtypes(include = 'number')

print('Categorical Dataframe')
display(C.head())
print('\n\nNumerical Dataframe')
display(D.head())
```

```python
# What is the average age of each Education Level.

o = mb[['education', 'age']]
o.groupby('education')[['age']].mean().sort_values(by = 'age', ascending = False)
```

```python
# Of all education level, what level has the most capital gain?

p = mb[['education', 'capital-gain']]
p.groupby('education')[['capital-gain']].sum().sort_values(by = 'capital-gain', ascending = False)
```

```python
# What race has the most capital loss

q = mb[['race', 'capital-loss']]
q.groupby('race')[['capital-loss']].sum().sort_values(by = 'capital-loss', ascending = False)
```

```python
# What is the total capital loss for all Female that is Divorced

r = mb[['gender', 'marital-status', 'capital-loss']]
r = r.loc[(mb.gender == 'Female') & (mb['marital-status'] == 'Divorced')]
r.groupby(['gender','marital-status'])[['capital-loss']].sum()
```

```python
# In your advice, if you are gonna give a loan, what Marital Status would you consider

s = mb[['marital-status', 'income_>50K']]
s.groupby(['marital-status'])[['income_>50K']].sum().sort_values(by = 'income_>50K', ascending = False).head(1)

#The one with the highest income?
```

```python
# SHow the total capital gain of all Doctorate education level

t = mb[['education', 'capital-gain']]
t = t.loc[(mb.education == 'Doctorate')]
t.groupby('education')[['capital-gain']].sum()
```

```python
# Who works the most total hours per week in the varying Occupation. What is the average capital gain of this class?

u = mb[['hours-per-week', 'occupation']]
u = u.groupby('occupation')[['hours-per-week']].mean().sort_values(by = 'hours-per-week', ascending = False).head(1)

uu = mb[['occupation', 'capital-gain']]
uu = uu.loc[mb.occupation == 'Farming-fishing']
uu = uu.groupby('occupation')[['capital-gain']].mean()

print('Who works the most total hours per week in the varying Occupation?')
display(u)
print('\n\nWhat is the average capital gain of this class?')
display(uu)
```

```python
# What race is divorces the most.
v = mb[['race', 'marital-status']]
v = v.loc[mb['marital-status'] == 'Divorced']
v.groupby('race')[['marital-status']].count().sort_values(by = 'marital-status', ascending = False)
```

```python
# What Marital Status works the most total hours. What race mostly constitute this status

w = mb[['marital-status', 'hours-per-week']]
w = w.groupby('marital-status')[['hours-per-week']].mean().sort_values(by = 'hours-per-week', ascending = False)

ww = mb[['race', 'marital-status']]
ww = ww.loc[ww['marital-status'] == 'Married-civ-spouse']
ww = ww.groupby('race')[['marital-status']].count().sort_values(by = 'marital-status', ascending = False)

print('What Marital Status works the most total hours?')
display(w)

print('\n\nWhat race mostly constitute this status')
display(ww)
```
